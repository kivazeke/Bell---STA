```python
x = np.array([1.25, 2, 3, 3.17, 2.13])    # <- Put your numbers here  Diameter
y = np.array([3.93, 6.28, 9.42, 9.96, 6.69])  #                           Circumference
```

```python
plt.plot(x,y,'g.')
```

```python
m_guess = 0
b_guess = 0
```

```python
m_guess * x + b_guess
```

```python
y_guess = m_guess * x + b_guess
```

```python
y_guess - y
```

```python
np.sum((y_guess - y)**2)
```

```python
loss = np.sum((y_guess - y)**2) 
```

```python
best_loss = 1000000000
best_m = 0
best_b = 0 

for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):

        print("Now Testing: ", m_guess, b_guess)
        
        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess
            best_b = b_guess 
```

```python
plt.title('Circle Diameter vs Circumference"')  # title

plt.xlabel('Diameter')  # x-axis label as Diameter
plt.ylabel('Circumference')  # y-axis label as Circumference

# data
plt.plot(x,y,'b.', label='5 Objects')  #plotting the data with blue dots

y_estimate = best_m * x + best_b # giving y_estimate a value

# line of best fit
plt.plot(x,y,'r--', label=f'Estimate: $y = {best_m} x + {best_b}$')

plt.legend();
```

```python
y_guess = best_m * x + best_b

plt.plot(x,y_guess,'r');
plt.plot(x,y,'g.');
plt.show()

y_estimate = best_m * x + best_b

plt.plot(x,y,'r--', label=f'Estimate: $y = {best_m} x + {best_b}$')

print(best_loss,best_m,best_b)
```
